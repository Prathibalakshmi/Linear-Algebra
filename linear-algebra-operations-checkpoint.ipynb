# Add noise to an image matrix

```python
mean = 0
sig = 3
noise_matr = mean + sig*np.random.randn(img0.shape[0],img0.shape[1])

noisy_img = img0 + noise_matr 
noisy_img = noisy_img.astype(np.uint8) ## uint8 -> unsigned 8-bit
plt.imshow(noisy_img,cmap="gray")
plt.imsave("noisy_sad_face.jpeg",noisy_img,cmap="gray")
```

## Norm of matrices and vectors

```python
from numpy import linalg as LA

A = np.array([3,-2,-1,-4])  
print("L1 norm for A is {}".format(LA.norm(A,1))) ## Calculates |3| + |-2| + |-1| + |4|, i.e., the L1 norm 
print("L2 norm for A is {}".format(LA.norm(A,2))) ## Calculates the L2 norm 
print("L-inf norm for A is {}".format(LA.norm(A,np.inf))) ## Calculates infinity norm, i.e. the maximum absolute value of the matrix

M = np.array([[1,2,3],
              [0,4,7]])
print("Frobenius norm for M is: {}".format(LA.norm(M,'fro'))) ## Frobenius norm
```

## Decomposition of a Matrix

```python
import scipy.linalg as la
A = np.array([[3,1,2],
              [1,6,9],
              [7,0,8]])
print(A)

P,L,U = la.lu(A)  ## P*L*U = A
print("The permutation matrix is {}".format(P))
print("The lower traingular matrix is {}".format(L))
print("The upper traingular matrix is {}".format(U))
X = np.matmul(L,U)
print(X)

A_new = np.matmul(P,X)
print(A_new)
```

# Some Useful Plots for Data Visualization

### Normal Contour Plot

```python
import numpy as np
import matplotlib.pyplot as plt
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
#Z = np.sqrt(X**2 + Y**2)
Z = 1.5*np.exp(-((X-1)**2 + (Y-1)**2)) + 2*np.exp(-((X-1)**2 + (Y+1)**2)) + 1.3*np.exp(-((X+2)**2 + (Y-1)**2)) # example function
fig,ax = plt.subplots(1,1,figsize=(9,6))
cp = ax.contour(X, Y, Z)
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Contour Plot')
plt.show()
```

### Filled Contour Plot

```python
import numpy as np
import matplotlib.pyplot as plt
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
#Z = np.sqrt(X**2 + Y**2)
Z = np.exp(-((X+1)**2 + (Y-1)**2)) + 1.5*np.exp(-((X-1)**2 + (Y-1)**2)) + 2*np.exp(-((X-1)**2 + (Y+1)**2))
fig,ax = plt.subplots(1,1,figsize=(9,6))
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Filled Contour Plot')
plt.show()
```

### 3-d Contour Plot

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x**2 + y**2

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(8,6))
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D contour')
plt.show()
```

### 3-d Surface Plot

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x**2 + y**2

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure(figsize=(9,6))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z)
ax.set_title('Surface plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
